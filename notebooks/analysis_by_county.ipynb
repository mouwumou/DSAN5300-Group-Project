```python
import pandas as pd
```

```python
df = pd.read_csv("../data/vehicle_registration_transactions.csv")
df.head()
```

```python
df['Fuel Type Primary'].value_counts()
```

```python
df['Year'] = pd.to_datetime(df['Transaction Month and Year']).dt.year
```

```python
df.isna().sum()
```

```python
df['Is Electric'] = df['Fuel Type Primary'].apply(lambda x: x == 'Electric')
df['Is Second Electric'] = df['Fuel Type Secondary'].apply(lambda x: x == 'Electric')
```

```python
df[df['Is Electric'] | df['Is Second Electric']]
```

```python
df['Make and Model'] = df['Make'] + ' ' + df['Model']
```

```python
grouped = df.groupby(['County', 'Year'])
```

```python
summary = grouped.agg(
    Total_Transactions=('Transaction Count', 'sum'),
    Electric_Vehicle_Count=('Is Electric', 'sum'),
    Electric_Vehicle_Ratio=('Is Electric', 'mean'),
    Second_Electric_Vehicle_Count=('Is Second Electric', 'sum'),
    Manufacturer_Distribution=('Make', lambda x: x.value_counts().to_dict()),
    Make_Model_Count=('Make and Model', 'nunique'),
    Make_Model_Distribution=('Make and Model', lambda x: x.value_counts().to_dict()),
    Vehicle_Type_Distribution=('Vehicle Type', lambda x: x.value_counts().to_dict()),
    Vehicle_Transaction_Type_Distribution=('Transaction Type', lambda x: x.value_counts().to_dict())
)
```

```python
summary['Electric_Vehicle_Ratio'] *= 100
```

```python
summary
```

```python
def get_electric_vehicle_type_distribution(x):
    electric_vehicle_mask = x['Is Electric'] | x['Is Second Electric']
    return x.loc[electric_vehicle_mask, 'Vehicle Type'].value_counts().to_dict()
```

```python
def get_electric_vehicle_model_distribution(x):
    electric_vehicle_mask = x['Is Electric'] | x['Is Second Electric']
    return x.loc[electric_vehicle_mask, 'Make and Model'].value_counts().to_dict()
```

```python
electric_vehicle_type_distribution = grouped.apply(get_electric_vehicle_type_distribution)
```

```python
electric_vehicle_model_distribution = grouped.apply(get_electric_vehicle_model_distribution)
```

```python
electric_vehicle_type_distribution
```

```python
# merge summary and two new dataframes

final_summary = pd.merge(summary.reset_index(), electric_vehicle_type_distribution.reset_index(name='Electric_Vehicle_Type_Distribution'), on=['County', 'Year'], how='left')
final_summary = pd.merge(final_summary, electric_vehicle_model_distribution.reset_index(name='Electric_Vehicle_Model_Distribution'), on=['County', 'Year'], how='left')
final_summary
```

```python
counties = final_summary['County'].unique()
years = [2020, 2021, 2022, 2023, 2024]
```

```python
all_combinations = pd.MultiIndex.from_product([counties, years], names=['County', 'Year'])
all_combinations_df = pd.DataFrame(index=all_combinations).reset_index()
all_combinations_df
```

```python
final_summary_full = pd.merge(all_combinations_df, final_summary, on=['County', 'Year'], how='left')
final_summary_full
```

```python
final_summary_full['Total_Transactions'] = final_summary_full['Total_Transactions'].fillna(0)
```

```python
final_summary_full.fillna({'Total_Transactions': 0, 
                           'Electric_Vehicle_Count': 0, 
                           'Electric_Vehicle_Ratio': 0, 
                           'Second_Electric_Vehicle_Count': 0, 
                           'Make_Model_Count': 0, 
                           }, inplace=True)
```

```python
dictionary_columns = ['Manufacturer_Distribution', 'Make_Model_Distribution', 'Vehicle_Type_Distribution', 
                      'Vehicle_Transaction_Type_Distribution', 'Electric_Vehicle_Type_Distribution', 'Electric_Vehicle_Model_Distribution']
for column in dictionary_columns:
    final_summary_full[column].fillna({i: {} for i in final_summary_full.index}, inplace=True)
```
