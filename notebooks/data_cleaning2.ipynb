### Now we fulfill the missing county information

- The Zipcode - County data is from website unitedstateszipcodes.org

```python
zip_df = pd.read_csv("../data/raw/zip_code_database.csv")
zip_df.head()
```

```python
def get_county(zip_code):
    zip_code = int(zip_code)
    try:
        county =  zip_df[zip_df['zip']==zip_code]['county'].values[0]
        county = county[:county.find(" County")]
    except:
        county =  "Unknown or Out of State"

    return county
```
